<font size="+2" color="indigo"><b>2. Операции сравнения</b></font><br>

Больше (>)

Меньше (<)

Проверка на равенство (==)

Проверка на неравенство (!=)

Больше или равно (>=)

Меньше или равно (<=)

**Примеры**

```python
7 > 5 # Заметим, что в операциях сравнения вывод всегда будет True или False
      # В данном случае ясно, что 7 больше 5, поэтому в выводе появилось True
```

```python
True, False # True и False обязательно писать с большой буквы, иначе возникнет ошибка
```

```python
true # Интерпретатор не понимает, что такое true, а вот True - понимает 
```

```python
false # Аналогичная ситуация, как и с True
```

```python
7 < 5 # В данном случае неверно, что 7 меньше 5, поэтому в выводе появилось False
```

```python
3 == 3 # Проверка на равенство. На равенство можно проверять любые объекты,
       # необязательно только числа
```

```python
3 = 3 # Так сравнивать нельзя, возникнет ошибка
```

```python
3 == 4 # В данном случаем понятно, что 3 не равно 4
```

```python
3 != 4  # Проверка на неравенство: 3 не равно 4 - следовательно, True
```

```python
3!=3 # Проверка на неравенство:3 равно 3 - следовательно, False
```

```python
3!3 # Не забываем ставить знак "равно"
```

```python
2 > 2, 2 >= 2  # Операция "больше или равно"
```

```python
2 > 2, 2 => 2 # В Python операцию "больше или равно" нельзя записывать
              # как в математике (=>), обязательно только так: >=
```

```python
3 < 3, 3 <= 3 # Операция "меньше или равно" записывается как в математике
```

```python
2 < 4 < 6, 2 <= 4 <= 6 # В Python можно делать даже такие сравнения, другие
                       # языки программирования такого не позволяют
```

```python
3 == 3 == 3 == 3 # И такое тоже возможно
```

```python
2 < 4 > 3 # В данном случае показано следующее: 4 больше 2 и больше 3 
```

<font size="+2" color="indigo"><b>3. Логические операции</b></font><br>

И (and)

ИЛИ (or)

НE  (not)

**Примеры**

```python
2 > 1 and 2 > 1 # 2 > 1 (True) and 2 > 1 (True) --> True
```

```python
1 > 2 and 2 > 1 # 1 > 2 (False) and 2 > 1 (True) --> False
```

```python
2 > 1 and 1 > 2 # 2 > 1 (True) and 1 > 2 (False) --> False
```

```python
1 > 2 and 1 > 2 # 1 > 2 (False) and 1 > 2 (False) --> False
```

```python
1 > 2 and 1 > 2 and 2 > 1  # 1 > 2 (False) and 1 > 2 (False) and 2 > 1 (True) --> False
```

```python
# Если используется операция "and", то общее выражение будет верным
# только в том случае, когда каждое из выражений будет истинным
```

```python
2 > 1 or 2 > 1 # 2 > 1 (True) or 2 > 1 (True) --> True
```

```python
1 > 2 or 2 > 1 # 1 > 2 (False) or 2 > 1 (True) --> True
```

```python
2 > 1 or 1 > 2 # 2 > 1 (True) or 1 > 2 (False) --> True
```

```python
1 > 2 or 1 > 2 # 1 > 2 (False) or 1 > 2 (False) --> False
```

```python
1 > 2 or 1 > 2 or 2 > 1  # 1 > 2 (False) or 1 > 2 (False) or 2 > 1 (True) --> True
```

```python
# Если используется операция "and", то общее выражение будет верным
# только в том случае, когда хотя бы одно из выражений будет истинным
```

```python
not True, not False # операция not конвертирует False в True, а True - в False
```

```python
not 7 > 6 # то, что 7 больше 6 - верно, 7 > 6 (True), not True --> False
```

```python
# Посмотрим на более сложное выражение
10 > 3 and 6 != 7 or 12 < 2 and not 13 > 12
# В логических выражениях сначало выполнется операция 1) not
# потом операция 2) and, а вконце уже 3) or
# Давайте подробнее рассмотрим вышеприведенную операцию
#10 > 3 and 6 != 7 or 12 < 2 and not 13 > 12
#        2         4         3    1          # В этой строке представлена
                                             # последовательность выполнения операций

# (1) not 13 > 12,   13 > 12 (True),   not True --> False
# (2) 10 > 3 (True) and 6 != 7 (True) --> True
# (3) 12 < 2 (False) and not 13 > 12 (False) --> False
# (4) ((10 > 3 and 6 != 7) (True) or (False) (12 < 2 and not 13 > 12)) --> True
```

```python
((10 > 3 and 6 != 7) or (12 < 2 and (not 13 > 12))) 
# Лучше расставлять скобки, так более читаемо, не нужно тратить 
# много времени на разбор взаимосвязей в сложном логическом выражении
```

**Простейший алгоритм (смена значений переменных)**

```python
# Даны две переменные со значениями, необходимо поменять
# значение этих переменных между собой.

# Реализация на любом другом языке программирования

a = 5       # Необходимо, чтобы a поменяло значение на 10
b = 10      # Необходимо, чтобы b поменяло значение на 5

tmp = a     # tmp = 5    
a = b       # a = 10, b = 10 
b = tmp     # b = 5, tmp = 5

a,b
```

```python
# Реализация этого же алгоритма на Python

a = 5
b = 10
a,b = b,a   # a, b = 10, 5
a,b  # До этого было a = 5, b = 10

# Как мы видим, дополнительная переменная не понадобилась
```

<font size="+2" color="indigo"><b>8. Создание библиотеки (площадь фигур)</b></font><br>

В прогроммировании повсеместно используются библиотеки. В библиотеках 
содержится код. Библиотеки ускоряют программирование, так как вам не нужно
заново писать алгоритмы, кто-то за вас их уже написал, вам остается только
установить библиотеку и найти необходиму функцию, которая реализует
нужный алгоритм.

Для того чтобы понять как устроены библиотеки, давайте попробуем создать свою, в 
которой будет содержаться код по расчёту площади фигур (круг, ромб, трапеция, треугольник)

```python
# Площадь ромба

print("start")
a = int(input("введите первую диогональ"))
b = int(input("введите вторую диогональ"))
S = (a*b)/2
print(S)
print("finish")
```

```python
# Площадь треугольника

a=int(input('Введите сторону треугольника'))
a
h=int(input('Введите высоту треугольника'))
h
S=1/2*a*h
S
```

```python
# Площадь круга

r = float(input('Введите радиус окружности: '))  
S = 3.14*(r**2) 
print('Площадь фигуры равна',S)
```

```python
# Площадь трапеции

a = int(input("Введите первое основание трапеции"))
b = int(input("Введите второе основание трапеции"))
c = int(input("Введите высоту трапеции")) 
S = ((a+b)/2)*c  
print(S)
```

Выше написан код для расчёта площади определенных фигур. 
Теперь необходимо поместить данный код в определенную структуру (в тело функции),
чтобы можно было использовать написанный код в качестве библиотеки.

```python
# Не обращаем внимание на операции "def" и "return", подробнее эти 
# операции изучим в будущем
# Каждый блок кода по расчёту фигур дополняем операциями "def" и "return"

def area_rhombus(): # Команда 4 # Площадь ромба
    
    print("start")
    a = int(input("введите первую диогональ"))
    b = int(input("введите вторую диогональ"))
    S = (a*b)/2
    print(S)
    print("finish")
    
    return S

def area_triangle(): # Команда 1 # Площадь треугольника
    
    a=int(input('Введите сторону треугольника'))
    a
    h=int(input('Введите высоту треугольника'))
    h
    S=1/2*a*h
    S

    return S

def area_circle(): # Команда 2 # Площадь круга
    
    r = float(input('Введите радиус окружности: '))
    S = 3.14*(r**2)
    print('Площадь фигуры равна',S)
    
    return S
     
def area_trapezoid(): # Команда 3 # Площадь трапеции
    
    a = int(input("Введите первое основание трапеции"))
    b = int(input("Введите второе основание трапеции"))
    c = int(input("Введите высоту трапеции"))
    S = ((a+b)/2)*c
    print(S)
    
    return S
```
