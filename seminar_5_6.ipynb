<font size="+2" color="indigo"><b>1. Сложные условия</b></font><br>

```python
# На прошлом занятии мы прошли тему "простые условия", однако на практике
# условия, состоящие из одного выражения, применяюются редко. 

# Пример простого условия показан ниже:

a = 5
if a > 2:
    print("Больше")
else:
    print("Меньше")
```

```python
# Сложное условие состоит из двух и более логических выражений.
# Давайте рассмотрим ниже пример сложного условия:

a = 5
b = 4
                           # (True) and (True)
if (a > 2 and b > 3):     # (a > 2 and b > 3)  --> True
    print("Условие верно") # Если условие верно (True) после оператора if, 
                           # то выполняется выражение после оператора if,                   
else:                      # а если False, то выполняется условие после else.
    print("Условие неверно")
```

```python
# Рассмотрим логическое выражение из трёх условий:

a = 5
b = 4
c = 6                                
                                     # (True) and (True)  or (False)
if ((a > 2 and b > 3) or c < 5):     # (a > 2 and b > 3)  or  c < 5   --> True
    print("Условие верно")
else:
    print("Условие неверно")
    
# Логическое выражение из трёх условий ничем не отличается от выражения с двумя условиями.
```

Тут-то и вступают в силу циклы. Благодаря им становится возможным максимально упростить данный вопрос" [https://python-scripts.com/loops-for-while].

<font size="+2" color="indigo"><b>2.1 Цикл while</b></font><br>

Цикл while выполняется, пока условие верно (True).

```python
# Цикл while начинается с оператора "while".
# После оператора "while" необходимо записать условие, 
# в конце необходимо поставить двоеточие, без этого будет ошибка. 
# Чтобы цикл выполнялся, необходимо записать тело цикла после строки с оператором "while".


count = 0

while count < 5:         # Пока count меньше пяти, цикл будет выполняться.

    count += 1     #count = count + 1 
    print(count)
    
#######################################
#            Первая итерация          #
#######################################

# count = 0
# while 0 < 5 (True):
#    count += 1   -->  count = 1 + 0 = 1
#    print(count) -->  1

#######################################
#            Вторая итерация          #
#######################################

# count = 1
# while 1 < 5 (True):
#    count += 1   -->  count = 1 + 1 = 2
#    print(count) -->  2

#######################################
#            Третья итерация          #
#######################################

# count = 2
# while 2 < 5 (True):
#    count += 1   -->  count = 2 + 1 = 3
#    print(count) -->  3

#######################################
#          Четвёртая итерация         #
#######################################

# count = 3
# while 3 < 5 (True):
#    count += 1   -->  count = 3 + 1 = 4
#    print(count) -->  4

#######################################
#            Пятая итерация           #
#######################################

# count = 4
# while 4 < 5 (True):
#    count += 1   -->  count = 4 + 1 = 5
#    print(count) -->  5

#######################################
#              Завершение             #
#######################################

# count = 5
# while 5 < 5 (False):  --> цикл больше не будет выполняться
```

<font size="+2" color="indigo"><b>2.2 Цикл for</b></font><br>

```python
# Цикл for начинается с оператора "for".
# После оператора "for" необходимо записать имя переменной, в которую
# будут записываться значения элементов списка (массива), по которому цикл итерируется.
# После этого необходимо записать генератор, массив или строку (объект, по элементам которого цикл будет итерироваться).
# В конце необходимо поставить двоеточие, без этого будет ошибка. 
# Чтобы цикл выполнялся, необходимо записать тело цикла после строки с оператором "for".
```

```python
for name in [12, 25, 56, 32]:  # Массив.
    print(name)
    
#######################################
#            Первая итерация          #
#######################################

# for name in [12, 25, 56, 32]: --> name = [12, 25, 56, 32][0]  --> name = 12
#     print(name) --> 12

#######################################
#            Вторая итерация          #
#######################################

# for name in [12, 25, 56, 32]: --> name = [12, 25, 56, 32][1]  --> name = 25
#     print(name) --> 25

#######################################
#            Третья итерация          #
#######################################

# for name in [12, 25, 56, 32]: --> name = [12, 25, 56, 32][2]  --> name = 56
#     print(name) --> 56

#######################################
#          Четвёртая итерация         #
#######################################

# for name in [12, 25, 56, 32]: --> name = [12, 25, 56, 32][3]  --> name = 32
#     print(name) --> 32
```

```python
# Если бы не было цикла, операция выше заняла бы четыре строчки. 
print([12, 25, 56, 32][0])
print([12, 25, 56, 32][1])
print([12, 25, 56, 32][2])
print([12, 25, 56, 32][3])
```

```python
for name in "Анализ данных":   # Строка.
    print(name)
```

```python
for name in range(0,10):       # Генератор.
    print(name)
```

```python
# Рассмотрим ещё один пример: с помощью цикла можно просуммировать все значения массива. 

count = 0

for i in [5, 5, 5, 5]:
    count +=i   
    print(count)   
    
print("Сумма значений массива = ",count)
```

<font size="+2" color="indigo"><b>2.3 Операторы continue, break</b></font><br>

Continue переводит цикл на следующую итерацию.

Break прерывает выполнение цикла.

```python
count = 0

while True:
    
    count += 1 
    if count < 10:    # Когда count будет равен 10, выполнится выражение, 
        print(count)
    else:             # которое идёт после оператора else,
        break         # то есть выполнится оператор break --> цикл прервётся.
                      # Без оператора break цикл был бы бесконечен. 
```

```python
# Пример использования оператора break в цикле for.

count = 0

for i in [5, 6, 7, 8]: # выполнится только одна итерация 
    
    count +=i   
    print(count) 
    break  
```

```python
count = 0

while True:
    
    count += 1 
    
    if count < 10:
        print(count)
        continue    # Строка ниже не будет выполняться из-за того,
                    # что оператор continue будет переводить цикл на следующую итерацию.
        count -= 5  # count = count - 1 
    else:
        break
```

```python
# Пример использования оператора continue в цикле for.

count = 0

for i in [5, 6, 7, 8]:
    
    print(count)
    continue
    count +=i 
```

<font size="+2" color="indigo"><b>3.1 Пример реализации алгоритмов</b></font><br>

Ниже представлена таблица с различными алгоритмами сортировки. 
Стандартная сортировка, которая представлена в Python, реализована на основе алгоритма Timsort.

Timsort и многие другие сортировки считаются быстрыми, и их реализация требует дополнительных знаний, поэтому мы рассмотрим четыре простые сортировки: глупая сортировка, сортировка пузырьком, шейкерная сортировка, чётно-нечётная сортировка.

<img src="https://habrastorage.org/storage1/61a272a8/54806d34/69954080/310f329f.png">

**Глупая сортировка**

<img style="-webkit-user-select: none;margin: auto;cursor: zoom-in;background-color: hsl(0, 0%, 90%);transition: background-color 300ms;" src="https://hsto.org/getpro/habr/post_images/7ae/81a/1aa/7ae81a1aa491bd3514e5dabdc68028ae.gif" width="4410" height="170">

```python
def sort_stupid(nums):
    
    swapped = True

    while swapped:

        swapped = False

        for i in range(len(nums)-1):
            
            if nums[i] > nums[i+1]:
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                swapped = True
                break
                
    return nums  
```

**Сортировка пузырьком**

<img style="-webkit-user-select: none;margin: auto;cursor: zoom-in;background-color: hsl(0, 0%, 90%);transition: background-color 300ms;" src="https://hsto.org/getpro/habr/post_images/187/5a3/929/1875a3929dd14c8ea5ff4ccc3d0db9bd.gif" width="4410" height="170">

```python
def sort_bubble(nums):
    
    swapped = True

    while swapped:

        swapped = False

        for i in range(len(nums)-1):
            
            if nums[i] > nums[i+1]:
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                swapped = True
                
    return nums
```

**Шейкерная сортировка**

<img style="-webkit-user-select: none;margin: auto;cursor: zoom-in;background-color: hsl(0, 0%, 90%);transition: background-color 300ms;" src="https://hsto.org/getpro/habr/post_images/2a9/ad7/855/2a9ad78556f13396ebc68cb4ac21e91c.gif" width="4410" height="170">

```python
def shaker_sort(nums): 
   
    swapped = True
    
    while swapped: 
        
        # Проход слева на право
        
        swapped = False
          
        for i in range(len(nums)-1):
            
            if (nums[i] > nums[i+1]):
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                swapped = True
  
        # если не было обменов прерываем цикл
        if (not(swapped)): 
            break

        swapped = False

  
        #проход справа налево
    
        for i in range(len(nums) - 2,  - 1, -1): 
            
            if (nums[i] > nums[i + 1]): 
                # обмен элементов
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                swapped = True
 
                
    return nums
```

**Чётно-нечётная сортировка**

<img style="-webkit-user-select: none;margin: auto;cursor: zoom-in;background-color: hsl(0, 0%, 90%);transition: background-color 300ms;" src="https://hsto.org/getpro/habr/post_images/d33/1e9/04f/d331e904f8681df245877c460ae6e6a8.gif" width="4410" height="170">

```python
def odd_even(nums):
    
    n = len(nums)
    isSorted = 0
    
    while isSorted == 0:
        
        isSorted = 1
        temp = 0
        
        for i in range(1, n - 1, 2):
            
            if nums[i] > nums[i + 1]:
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                isSorted = 0
        for i in range(0, n - 1, 2):
            
            if nums[i] > nums[i + 1]:
                nums[i], nums[i + 1] = nums[i + 1], nums[i]
                isSorted = 0
                
    return nums
```

<font size="+2" color="indigo"><b>3.2 Скорость выполнения алгоритмов</b></font><br>

```python
# Сгенерируем массив из 10_000 элементов. 

import numpy as np
import random
nums = np.array([random.random() for i in range(10000)])
```

```python
# Скопируем массив, так как в процессе сортировки первоначальный массив изменится.
# Чтобы сравнить скорость алгоритмов, нам необходимо четыре одинаковых массива. 

nums_sort_stupid = nums.copy()
nums_sort_bubble = nums.copy()
nums_shaker_sort = nums.copy()
nums_odd_even = nums.copy()
```

```python
%%time
sort_stupid(nums_sort_stupid) # Нет смысла выполнять, так как примерно 
                              # алгоритм совершит 1_000_000_000_000 (триллион) операций.
```

```python
%%time
sort_bubble(nums_sort_bubble)
```

```python
%%time
shaker_sort(nums_shaker_sort)
```

```python
%%time
odd_even(nums_odd_even)
```

```python
%%time
nums_sorted = nums.copy()
sorted(nums_sorted)

# Стандартная сортировка в разы лучше по скорости выполнения,
# чем алгоритмы расмотренные выше.
```

```python
sorted(nums_sorted, reverse = True) # Если необходимо отсортировать значения 
                                    # в обратном порядке, то используем reverse = True
```
